## Choosing an appropriate algorithm

<img src="algorithms.png" width=1400 />

1.We cannot represent the problem with a table case since the major features of this problem are continuous, so let's use function approximation.

2.We will not use average reward since this is an episodic task.

3.We will learn at each time step, so we are incorporating TD(0) instead of MC methods, since, for example, the left motor can break down in the middle of the land or the agent can get out of fuel.

4.Next, is this a control problem or a predict problem? This is a control problem since we want to optimize our policy via policy-based methods.

5.After this, we conclude our search with the following algorithms: Q-Learning, Sarsa, Expected Sarsa. Note that this is a mere example. We have so many more algorithms to cover here, but it is an extensive list, so let's remain with the basic ones.

6.Q-Learning uses a deterministic policy in the target function of the TD error. Since we want to favor exploration, let's use an e-greedy sub-optimal policy. We remain with Sarsa and Expected Sarsa. We know that Expected Sarsa gives better results than Sarsa, because it makes the backup updates with the expected value instead of the sample value like Sarsa does. In contrast, it takes more time. Nevertheless, let's remain with Expected Sarsa.
