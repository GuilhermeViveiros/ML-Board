## Solving the Mountain Cair with Reinforcement Learning

> About the algorithm:

    1. Temporal Difference Learning - TD(0) to resolve a finite MDP
    2. Episodic Tasks
    3. Discounted returns
    4. On-Policy
    5. Model-Free

```python
class QLearning():
    
    #Initialize the agent
    def agent_init(self, agent_info = {}):
        raise NotImplementedError
    
    #Start the episodic task, with a stochastic action follwing a arbitrary policy
    def agent_start(self,state):
        raise NotImplementedError
    
    #When the agent finds the terminal state
    def agent_end(self):
        raise NotImplementedError
    
    #Take an action and observed the current state
    #Update the value function
    def agent_step(self,state,reward):
        raise NotImplementedError
        
    #Argmax function with random ties
    def argmax(self, actions):
        ties = []
        max_ = float("-inf")
        
        for i in range(len(actions)):
            if(actions[i] > max_):
                ties = []
                max_ = actions[i]
                
            if(actions[i] == max_):
                ties.append(i)
        
        return self.rand_generator.choice(ties)
```

```python
%%add_to TD_Agent

#Following a greedy policy with e = self.epsilon
def agent_init(self, agent_info = {}):
    #actions
    self.num_actions = agent_info.get("num_actions")
    #states
    self.num_states = agent_info.get("num_states")
    #discount
    self.discount = agent_info.get("discount")
    #step_size
    self.step_size = agent_info.get("step_size")
    #epsilon
    self.epsilon = agent_info.get("epsilon")
    #random generator
    self.rand_generator = np.random.RandomState(agent_info.get("seed",3))
    #Action value function
    self.q = np.zeros(shape=(self.num_states,self.num_actions))
```

```python
current_q = np.zeros(shape=(2400,4))
```

```python
%%add_to TD_Agent

def agent_start(self,state):
    #I'm following a greedy action so
    current_q = self.q[state]
    #stochastic
    if (self.rand_generator.rand() < self.epsilon):
        action = self.rand_generator.randint(self.num_actions)
    #greedy
    else:
        action = self.argmax(current_q)
    
    self.prev_action = action
    self.prev_state = state
    
    return action
```

```python
%%add_to TD_Agent

def agent_step(self,state,reward):
    

    current_q = self.q[state]

    #I'm following a greedy action so
    #stochastic
    if (self.rand_generator.rand() < self.epsilon):
        action = self.rand_generator.randint(self.num_actions)
    #greedy
    else:
        action = self.argmax(current_q)
    
    # Perform an update
    target = reward + self.discount * np.max(self.q[state])
    self.q[self.prev_state][self.prev_action] = self.q[self.prev_state][self.prev_action] + self.step_size * (target - self.q[self.prev_state][self.prev_action])
    
    # Update the states
    self.prev_state = state
    self.prev_action = action
    
    return action
```

```python
%%add_to TD_Agent

def agent_end(self,reward):
    self.q[self.prev_state][self.prev_action] = self.q[self.prev_state][self.prev_action] + self.step_size * (reward - self.q[self.prev_state][self.prev_action])
```

## Environment and the Agent done

> Now let's play some finite episodes with open-ai, and see the results for finding an optimal policy with Q-Learning

```python
game = 0

#one hundred generations
mean_rewards_per_game = []


eps_decay = 1/40000

while game < 40000:
    
    if(game%50 == 0):
        print("Generated game = {0}".format(game))
        
    
    game+=1
    rewards = []
   
    state = model.env_start()
    
    #Every state that goes to the agent, first need to be encoded to is index
    state = model.state_index((state[0],state[1]))
    action = agent.agent_start(state)
    
    (reward, state, done) = model.env_step(action)
    rewards.append(reward)
    state = model.state_index((state[0],state[1]))
    
    
    #max 200 iterations
    while(done is not True):
        #If wonna render the game and check the training progress de-coment the next two lines (It will take longer to train)
        #if(game % 50 == 0):
        #    model.env.render()
        action = agent.agent_step(state,reward)
        (reward, state, done) = model.env_step(action)
        state = model.state_index((state[0],state[1]))
        
        #stored information
        rewards.append(reward)
                
                
    #if(np.sum(np.asarray(rewards)) > -200):
    #    print(np.sum(np.asarray(rewards)))
    #    print("Successfully reached the termination state")
    
    mean_rewards_per_game.append(np.sum(rewards))
                    
    agent.epsilon = agent.epsilon - eps_decay
    
        
        
            
```
